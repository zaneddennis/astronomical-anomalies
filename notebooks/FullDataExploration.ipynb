```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
preprocessed_data = np.load('fulldata.npz',allow_pickle=True)
times = preprocessed_data["times"]
fluxes = preprocessed_data["fluxes"]
flux_errs = preprocessed_data["flux_errs"]
filters = preprocessed_data["filters"]
classes = preprocessed_data["ids"]
```

```python
t_lens = []
for e_times in times:
    t_lens.append(len(e_times))

t_lens
```

```python
print(min(t_lens))
print(max(t_lens))
```

```python
plt.hist(t_lens, bins=30)
```

```python
len(np.where(np.array(t_lens)>=10)[0])
```

```python
toKeep = np.where(np.array(t_lens)>=10)[0]

times = times[toKeep]
fluxes = fluxes[toKeep]
flux_errs = flux_errs[toKeep]
filters = filters[toKeep]
classes = classes[toKeep]
```

```python
np.savez("fulldata_longcurves.npz", times=times, fluxes=fluxes, flux_errs=flux_errs, filters=filters, ids=classes)
```

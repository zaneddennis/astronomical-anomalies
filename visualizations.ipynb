```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = "notebooks/fulldata.npz"
to_analyze = 40
NUM_PASSBANDS = 6
```

```python
def plot_event(e_times, e_fluxes, e_filters, event_id):
    for passband, color in zip(range(NUM_PASSBANDS), ("blue", "cyan", "green", "yellow", "orange", "red")):
        good_ixes = np.where(e_filters == passband)
        plt.plot(e_times[good_ixes], e_fluxes[good_ixes], color=color, marker="o")

    plt.ylim(-0.1, np.amax(e_fluxes)+0.5)
    plt.xlabel("Time (days)")
    plt.ylabel("Flux")
    plt.title("Light Curve {}".format(event_id))
    plt.show()
```

```python
data = np.load(filepath, allow_pickle=True)
times = data["times"][:to_analyze]
fluxes = data["fluxes"][:to_analyze]
flux_errors = data["flux_errs"][:to_analyze]  # unused so far
filters = data["filters"][:to_analyze]
classes = data["ids"][:to_analyze]
```

```python
for i in range(to_analyze):
    e_times = times[i]
    e_fluxes = fluxes[i]
    e_filters = filters[i]
    
    plot_event(e_times, e_fluxes, e_filters, i)
```

```python
a_times = times[0]
a_fluxes = fluxes[0]
a_filters = filters[0]
b_times = times[11]
b_fluxes = fluxes[11]
b_filters = filters[11]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

for passband, color in zip(range(NUM_PASSBANDS), ("blue", "cyan", "green", "yellow", "orange", "red")):
    good_ixes = np.where(a_filters == passband)
    ax1.plot(a_times[good_ixes], a_fluxes[good_ixes], color=color, marker="o")

#plt.ylim(-0.1, np.amax(a_fluxes)+0.5)
#plt.xlabel("Time (days)")
#plt.ylabel("Flux")
#plt.title("Light Curve 0")
ax1.axis(ymin=-0.1, ymax=np.amax(a_fluxes)+0.5)
ax1.set(xlabel="Time (days)", ylabel="Flux", title="Light Curve 0")

for passband, color in zip(range(NUM_PASSBANDS), ("blue", "cyan", "green", "yellow", "orange", "red")):
    good_ixes = np.where(b_filters == passband)
    ax2.plot(b_times[good_ixes], b_fluxes[good_ixes], color=color, marker="o")

#plt.ylim(-0.1, np.amax(b_fluxes)+0.5)
#plt.xlabel("Time (days)")
#plt.ylabel("Flux")
#plt.title("Light Curve 11")
ax2.axis(ymin=-0.1, ymax=np.amax(b_fluxes)+0.5)
ax2.set(xlabel="Time (days)", ylabel="Flux", title="Light Curve 11")

plt.show()
```
